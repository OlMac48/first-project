—Å—Ç–æ–∏—Ç –æ–±—Ä–∞—Ç–∏—Ç—å –≤–Ω–∏–º–∞–Ω–∏–µ –Ω–∞ –ø–µ—Ä–≤—ã–π –∏ –Ω–∞ —á–µ—Ç–≤–µ—Ä—Ç—ã–π    
    
    

```python
# import pandas as pd
n_rows = 3
result = pd.DataFrame(
    {"a": [list(range(1 + i ** 2)) for i in range(n_rows)], "b": list(range(n_rows))}
)
```

```python
result
```

```python
result.loc[3]=([4, 2, 1, 4, 3], 2)
result
```

```python
result = result.explode("a").astype({'a':int})
result
```
